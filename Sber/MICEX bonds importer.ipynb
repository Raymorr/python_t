### Calculate turnover for individual security

##### Sber EQT example of turnover calculation

```python
#download data

Sber = '''
    select * 
    from tickdata.trades t 
    where (t.tradedate = '20201201') 
    AND 
    (t.symbol like '%%MICEX.EQT.SBER.M2%%')''' #here you can paste any security you want (in case you have its greenplum name)

sql = \
Sber

 
# read a table from database into pandas dataframe, replace "tablename" with your table name
Sber_table = pd.read_sql(sql,engine)
Sber_table.head()
```

```python
# sber turnover calculation

Sber_table['turnover'] = Sber_table['price']*Sber_table['size']
a = round(Sber_table['turnover'].sum()/1000000000,1) #bn rub overall turnover per day

Sber_table['local_timestamp'] = Sber_table['local_timestamp'].apply(lambda x: str(x))
Sber_table1 = Sber_table['18:50:00' < Sber_table['local_timestamp']]
Sber_table1['turnover'] = Sber_table1['price']*Sber_table1['size']
b = round(Sber_table1['turnover'].sum()/1000000000,1) #bn rub turnover evening trading session
```

```python
# evening trading session turrnover / total turnover

print('Evening trading turnover constitues',round(b/a*100,1),'% of the total turnover per day')
```

```python
#Sber EQT trading regimes

Sber_table['Regime'] = (Sber_table['symbol'].str.extract(r'\b(\w+)$'))
Sber_table['turnover'] = Sber_table['price']*Sber_table['size']
Sber_table.groupby(['Regime']).sum()
```

```python
#Sber daily turnover in RUB

res = int(Sber_table.groupby(['Regime']).sum().turnover)
print("Sber daily turnover is",f"{res:.0f}", "RUB")
```
